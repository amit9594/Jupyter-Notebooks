```python
def length(s):
    print(len(s))

def uppercase(s):
    print(s.upper())

def lowercase(s):
    print(s.lower())

def split(s):
    At_what = input("Where You want to split? so enter a character or if don't want then simply use space button: ")
    print(s.split(At_what))

def first_capital(s):
    print(s.capitalize())

def replace(s):
    where = input("Where You want to replace? ")
    what = input("By which letters You want to replace? ")
    print(s.replace(where,what))

def find(s):
    what = input("What you want to find, Please Enter that sub-string: ")
    print(f'{what}, this sub-string is present at index: {s.find(what)}')

def count(s):
    letter = input("Enter a character/sub-string which you want to find out its count: ")
    print(s.count(letter))
    
def string():
    print('*****Menu*****')
    print('1.Find Length of string')
    print('2.Make Uppecase')
    print('3.Make lowecase')
    print('4.Split string')
    print('5.Make first letter capital')
    print('6.Replace characters or sub-string by any other string: ')
    print('7.Find a sub-string: ')
    print('8.Find count of particular character/sub-string')




    choice = 'wrong'
    within_range = False
    acceptable_range = range(1,9)
    while choice.isdigit() == False or within_range == False:
        choice = input("Please Enter option in (1,8): ")
        
        if choice.isdigit() == False:
            print("Sorry this is not digit, please enter a valid digit: ")
        
        if choice.isdigit() == True:
            if int(choice) in acceptable_range:
                within_range = True
            else:
                print("Sorry, You are out of range, please enter from (1,8): ")
                within_range = False
    
    choice = int(choice)  

    if choice == 1:
        length(input("Enter string: "))
    elif choice == 2:
        uppercase(input("Enter string: "))
    elif choice == 3:
        lowercase(input("Enter string: "))
    elif choice == 4:
        split(input("Enter string: "))
    elif choice == 5:
        first_capital(input("Enter string: "))
    elif choice == 6:
        replace(input("Enter string: "))
    elif choice == 7:
        find(input("Enter string: "))
    elif choice == 8:
        count(input("Enter string: "))

```

```python
string()
```
